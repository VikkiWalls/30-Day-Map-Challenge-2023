## Day 19 - 5 minute map

Plan: Use OpenStreetMap to map an area as fast as I can

```python
import osmnx as ox
```

Selecting roads of Paris using coordinates (obtained by viewing OpenStreetMap Export page)

```python
Streets = ox.graph_from_bbox(48.8904,48.8553,2.2639,2.3249,retain_all=True, truncate_by_edge=True,
                        simplify=False, network_type='drive')
```

Plotting the map

```python
fig, ax = ox.plot_graph(Streets,figsize=(50,50), edge_color='red', bgcolor='black', node_size=0.1)
```
