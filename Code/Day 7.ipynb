```python
import os
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

%matplotlib inline
px.defaults.width = 1090
px.defaults.height = 600
```

Defining the start and end locations

```python
route = pd.read_csv("NY to Amarillo Route.csv")
route.head()
```

```python
px.set_mapbox_access_token(open('mapboxtoken.txt').read())
```

```python
fig = px.line_mapbox(route, lat="Lat", lon="Long")
fig.add_trace(go.Scattermapbox(
    mode = "lines",
    lon = route.iloc[:,1],
    lat = route.iloc[:,0],
    marker = dict(size= 10, color = '#FA9B6A'),
    name = "The way to Amarillo"))

fig.update_layout(mapbox_style="mapbox://styles/mapbox/dark-v11", mapbox_zoom=4.3, mapbox_center_lat = 39,mapbox_center_lon = -89,
    margin={"r":0,"t":0,"l":0,"b":0},paper_bgcolor = '#232222',legend_font_color="#FFF998")

fig.show()
```

```python
fig.write_html("navplot.html")
```
