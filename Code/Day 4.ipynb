```python
import os
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

%matplotlib inline
px.defaults.width = 1300
px.defaults.height = 800
```

Importing the [data](https://github.com/VikkiWalls/30-Day-Map-Challenge-2023/tree/main/Other%20Files/Hand%20Collated%20Data)

_All of the data was sourced from [Wikipedia](https://en.wikipedia.org/wiki/List_of_mass_shootings_in_the_United_States_in_2023#cite_note-272)_

```python
Shootings = pd.read_csv("2023 mass shootings USA.csv")
Shootings.head()
```

Importing my mapbox token

```python
px.set_mapbox_access_token(open('mapboxtoken.txt').read())
```

Building the map

```python
fig = px.scatter_mapbox(Shootings, lat="Lat", lon="Long")
fig.add_trace(go.Scattermapbox(
    lon = Shootings.iloc[:,1],
    lat = Shootings.iloc[:,0],
    marker = dict(size= 8, color = '#ff1100'),
    name = "Mass Shootings"))

fig.update_layout(mapbox_style="mapbox://styles/mapbox/dark-v11", mapbox_zoom=3, mapbox_center_lat = 35,mapbox_center_lon = -113,
    margin={"r":0,"t":0,"l":0,"b":0},paper_bgcolor = 'rgb(31, 31, 31)',legend_font_color="red")


fig.show()
```

```python
fig.write_html("badplot.html")
```
