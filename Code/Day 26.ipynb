# 30 Day Map Challenge

## Day 26 - Minimal

Plan: Plot a minimalist road map of Verona, Italy

Setting up the Notebook:

```python
import osmnx as ox
```

Checking Osmnx Version

```python
ox.__version__
```

Setting the location

```python
place = "Verona, Italy"
```

Defining the Geo Data Frame

```python
gdf = ox.geocode_to_gdf(place)
```

Telling Osmnx what type of data to use (i.e. roads, not pavements etc)

```python
G = ox.graph_from_place(place, network_type="drive", retain_all=True)
```

Plotting the map

```python
fig, ax = ox.plot_graph(
    G,
    #show=True,
    close=True,
    bgcolor="#ffffff",
    edge_color="#03092b",
    edge_linewidth=2.5,
    node_size=0,
    figsize=(50,50)
)
```
